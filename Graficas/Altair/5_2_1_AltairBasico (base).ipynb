# Grammar of Graphics
image.png

**Altair** se basa en el concepto fundamental llamado **La Gramtica de Grficos**. Esta divide cada visualizacin en:
* *Datos*: qu informacin se est pasando a la visualizacin?
* *Esttica*: Qu elementos visuales vemos como parte de esa visualizacin?
* *Escalas*: Cmo mapeamos de los datos a lo visual?
* *Objetos geomtricos*: estamos usando lneas? Estamos usando barras? Estamos usando puntos?
* *Estadsticas*: Estamos mostrando los datos en bruto? Estamos mostrando promedios? Estamos mostrando regresiones o tendencias?
* *Facetas* en las que dividimos nuestros datos, qu dimensiones o subconjuntos elegimos mostrar?
* *Sistemas de coordenadas*: Tpicamente hablando, estos van a ser nuestras coordenadas cartesianas lineales o coordenadas polares, aunque ocasionalmente vers ejemplos que utilizan sistemas de coordenadas menos convencionales.

La razn por la que nos enfocamos en la gramtica de grficos es que subyace es la base de varios kits de herramientas de visualizacin comunes, incluyendo Altair.

# Altair

image.png

No hay tantas limitaciones, pero algunos grficos pueden ser difciles de armar (por ejemplo, no hay 3D)

**Altair es declarativo**

Esto significa que se tiene que declarar qu se quiere, y Altair se encarga de los detalles. Hay tres parmetros importantes:

1. **` Marks: `** Cmo quieres que se muestren los datos? Con lineas, puntos, crculos, barras?
2. **`Channels:`** Qu debera ir en el eje x? Qu columna define los colores?
3. **`Encoding:`** Los datos son una fecha? Nmeros? Categoras?

(y otros... como *transformations*, *interactions*, *compounding*)

## Primer ejemplo

```python
#Datos:
import pandas as pd
from pandas import DataFrame

data=pd.DataFrame({'id_pais':[1, 2, 3, 4, 5, 6],
                    'poblacion':[1, 100, 200, 300, 400, 500],
                    'ingreso':[50, 50, 200, 300, 300, 450]})
print(data.info())
print(data)
```

**Sintaxis bsica**

image.png

```python
import altair as alt

# Transcribe el cdigo de la imagen

categorical_chart = alt.Chart(data).mark_circle(size=200).encode(
    x = 'poblacion:Q',
    color = 'id_pais:Q')
```

```python
categorical_chart
```

## Otras caractersticas bsicas
### Tipos de datos
Qu pasa si cambiamos el tipo de dato?. Altair maneja los siguientes tipos de datos:

* Cuantitativo (Q)

* Ordinal (O)

* Nominal (N)

* Temporal (T)

```python
# cambio de tipo de dato a Ordinal
chart2 = alt.Chart(data).mark_circle(size=200).encode(
    x = 'poblacion:Q',
    color = 'id_pais:O')
chart2
```

```python
# 2D scatter plot con tooltip
chart3 = alt.Chart(data).mark_circle(size=200).encode(
    x = 'poblacion:Q',
    color = 'id_pais:N')
chart3
```

### Tamao del grfico

```python
# Propiedades del grfico
print(chart3.to_json(indent=2))
```
