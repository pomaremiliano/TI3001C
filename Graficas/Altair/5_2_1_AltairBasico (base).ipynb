# Grammar of Graphics
image.png

**Altair** se basa en el concepto fundamental llamado **La Gramática de Gráficos**. Esta divide cada visualización en:
* *Datos*: ¿qué información se está pasando a la visualización?
* *Estética*: ¿Qué elementos visuales vemos como parte de esa visualización?
* *Escalas*: ¿Cómo mapeamos de los datos a lo visual?
* *Objetos geométricos*: ¿estamos usando líneas? ¿Estamos usando barras? ¿Estamos usando puntos?
* *Estadísticas*: ¿Estamos mostrando los datos en bruto? ¿Estamos mostrando promedios? ¿Estamos mostrando regresiones o tendencias?
* *Facetas* en las que dividimos nuestros datos, ¿qué dimensiones o subconjuntos elegimos mostrar?
* *Sistemas de coordenadas*: Típicamente hablando, estos van a ser nuestras coordenadas cartesianas lineales o coordenadas polares, aunque ocasionalmente verás ejemplos que utilizan sistemas de coordenadas menos convencionales.

La razón por la que nos enfocamos en la gramática de gráficos es que subyace es la base de varios kits de herramientas de visualización comunes, incluyendo Altair.

# Altair

image.png

No hay tantas limitaciones, pero algunos gráficos pueden ser difíciles de armar (por ejemplo, no hay 3D)

**Altair es declarativo**

Esto significa que se tiene que declarar qué se quiere, y Altair se encarga de los detalles. Hay tres parámetros importantes:

1. **` Marks: `** ¿Cómo quieres que se muestren los datos? ¿Con lineas, puntos, círculos, barras?
2. **`Channels:`** ¿Qué debería ir en el eje x? ¿Qué columna define los colores?
3. **`Encoding:`** ¿Los datos son una fecha? ¿Números? ¿Categorías?

(y otros... como *transformations*, *interactions*, *compounding*)

## Primer ejemplo

```python
#Datos:
import pandas as pd
from pandas import DataFrame

data=pd.DataFrame({'id_pais':[1, 2, 3, 4, 5, 6],
                    'poblacion':[1, 100, 200, 300, 400, 500],
                    'ingreso':[50, 50, 200, 300, 300, 450]})
print(data.info())
print(data)
```

**Sintaxis básica**

image.png

```python
import altair as alt

# Transcribe el código de la imagen

categorical_chart = alt.Chart(data).mark_circle(size=200).encode(
    x = 'poblacion:Q',
    color = 'id_pais:Q')
```

```python
categorical_chart
```

## Otras características básicas
### Tipos de datos
¿Qué pasa si cambiamos el tipo de dato?. Altair maneja los siguientes tipos de datos:

* Cuantitativo (Q)

* Ordinal (O)

* Nominal (N)

* Temporal (T)

```python
# cambio de tipo de dato a Ordinal
chart2 = alt.Chart(data).mark_circle(size=200).encode(
    x = 'poblacion:Q',
    color = 'id_pais:O')
chart2
```

```python
# 2D scatter plot con tooltip
chart3 = alt.Chart(data).mark_circle(size=200).encode(
    x = 'poblacion:Q',
    color = 'id_pais:N')
chart3
```

### Tamaño del gráfico

```python
# Propiedades del gráfico
print(chart3.to_json(indent=2))
```
