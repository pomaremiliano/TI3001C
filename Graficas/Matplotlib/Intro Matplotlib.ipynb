```python
import numpy as np
x = np.linspace(0, 2*np.pi, 100) # 100 puntos entre 0 y 2pi

# 1er paso: crear "canvas"
plt.figure()

# especificar posici√≥n
plt.subplot(2, 1, 1) # 2 filas, 1 columna, 1er panel
# agregar un primer panel
plt.plot(x, np.sin(x)) # gr√°fico de la funci√≥n seno

# especificar posici√≥n del 2do panel
plt.subplot(2, 1, 2) # 2 filas, 1 columna, 2do panel
# agregar un segundo panel
plt.plot(x, np.cos(x), 'g--') # gr√°fico de la funci√≥n coseno
plt.show() # muestra el gr√°fico
```

```python
import pandas as pd

df = pd.read_csv('pune_1965_to_2002.csv').set_index('Year')
df
```

```python
x = df.index # a√±os
y = df['May'] # precipitacion en Mayo

fig, ax = plt.subplots() # crear canvas y 1 panel
ax.plot(x, y, 'r--')
ax.set_xlabel('A√±o')
ax.set_ylabel('Precipitaci√≥n (mm)')
ax.set_title('Precipitaci√≥n de Mayo entre 1965-2002')
plt.show()
```

```python
x = df.index # a√±os
y = df.sum(axis='columns') # suma de precipitaciones por a√±o
# figsize=(6, 2): # tama√±o del gr√°fico
fig, ax = plt.subplots(figsize=(8,3)) # crear canvas y 1 panel
ax.plot(x, y, 'b', label='Precipitaci√≥n total')

ax.set_xlabel('A√±o')
ax.set_ylabel('Precipitaci√≥n (mm)')
ax.set_title('Precipitaci√≥n acumulada entre 1965-2002')
ax.legend(loc='lower right') # leyenda, loc: ubicaci√≥n de la leyenda

ax.set_xlim(df.index[0], df.index[-1])
ax.set_ylim(y.min(), y.max()) # que podemos utilizar para cambiar el rango de los ejes verticales?
plt.show()
```

```python
fig, ax = plt.subplots()  # crear panel
mapeo_colores = ax.scatter(df['Aug'], df['Sep'], c=df.index)
fig.colorbar(mapeo_colores) # agregar barra de colores
ax.set_title('Precipitaciones Agosto-Septiembre')
ax.set_xlabel('Precipitaciones en Agosto (mm.)')
ax.set_ylabel('Precipitaciones en Septiembre (mm.)')
plt.show()
```

## Gr√°ficos de barras

```python
# X: categorica
# Y: numerica
fig, ax = plt.subplots(figsize=(10,5)) # crear canvas y 1 panel
y = df.sum() # suma de precipitaciones por mes
ax.bar(df.columns, y)
ax.set_title('Precipitaciones por mes')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,3)) # crear canvas y 1 panel
ax.plot(df.index, df['Jan'], label='Precipitaciones de enero')
ax.plot(df.index, df['Feb'], label='Precipitaciones de febrero', color='C1')

maximo_enero = df['Jan'].max()
maximo_febrero = df['Feb'].max()

ax.axhline(maximo_enero, color='red', linestyle='--', alpha=0.5, linewidth=3, label='M√°xima de enero')
ax.axhline(maximo_febrero, color='red', linestyle=':', alpha=0.5, linewidth=3, label='M√°xima de febrero')

ax.set_xlabel('A√±o')
ax.set_ylabel('Precipitaci√≥n (mm.)') 
ax.set_title('Precipitaciones de enero y febrero') 	
ax.set_xlim(df.index[0], df.index[-1])
ax.legend()

ind_max = df['Jan'].idxmax() # √≠ndice del m√°ximo: a√±o
ax.axvline(ind_max, color = 'green', linewidth=2, alpha=0.5) # l√≠nea vertical
ax.annotate(f'En {ind_max} llovi√≥ mucho', xy=(ind_max, 4),
            xytext=(ind_max - 8, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(df.values.flatten(), bins=10) # bins: n√∫mero de intervalos
ax.set_title('Histograma de precipitaciones')
ax.set_xlabel('Intervalos de precipitaciones (mm.)')
ax.set_ylabel('Frecuencia absoluta') # el conteo de observaciones en cada bin
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(df.values.flatten(), bins=10, density=True) # bins: n√∫mero de intervalos
ax.set_title('Histograma de precipitaciones')
ax.set_xlabel('Intervalos de precipitaciones (mm.)')
ax.set_ylabel('Frecuencia relativa') # el porcentaje de observaciones en cada bin
plt.show()
```

## Boxplots

```python
fig, ax = plt.subplots(figsize=(8, 2))
ax.boxplot(df.values.flatten(), orientation='horizontal') # bins: n√∫mero de intervalos
ax.set_title('Diagrama de cajas de precipitaciones')
plt.show()
```

## Gr√°ficas de pastel

```python
fig, ax = plt.subplots()
ax.pie(df.sum(), labels=df.columns, autopct='%1.1f%%') # autopct: porcentaje
ax.set_title('Porcentaje de precipitaciones por mes')
plt.show()
```
