## 2. (10 pts) Con el mismo dataframe de salarios, genera una figura con 4 sub gráficos de distinto tipo mostrando diferentes comparativas y desagregando (hue) por alguna variable categórica. Acompaña con un bloque de texto con tu interpretación y justificación para cada subgráfico.

```python
fig, axs = plt.subplots(2, 2, figsize=(16, 12))

# Boxplot de salarios por departamento
sns.boxplot(data=df, x='Department', y='Salary', hue='Sex', ax=axs[0, 0])
axs[0, 0].set_title('1. Salario por Departamento y Sexo')
axs[0, 0].set_xticklabels(axs[0, 0].get_xticklabels(), rotation=45)

# Histograma de ausencias por satisfacción laboral
sns.histplot(data=df, x='Absences', hue='EmpSatisfaction', multiple='stack', bins=10, ax=axs[0, 1])
axs[0, 1].set_title('2. Ausencias por Nivel de Satisfacción')

# Violinplot de EngagementSurvey por Departamento
sns.violinplot(data=df, x='Department', y='EngagementSurvey', hue='Sex', split=True, ax=axs[1, 0])
axs[1, 0].set_title('3. Engagement Survey por Departamento y Sexo')
axs[1, 0].set_xticklabels(axs[1, 0].get_xticklabels(), rotation=45)

# Countplot de empleados por fuente de reclutamiento y sexo
sns.countplot(data=df, x='RecruitmentSource', hue='Sex', ax=axs[1, 1])
axs[1, 1].set_title('4. Empleados por Fuente de Reclutamiento y Sexo')
axs[1, 1].set_xticklabels(axs[1, 1].get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()
```

- Gráfica 1: El departamento que más persibe es el de IT/IS y el de Software Engineering. Justamente en esos dos departamentos es donde se observa una brecha más grande entre hombres y mujeres.

- Gráfica 2: Se aprecia que los empleados con menor satisfacción tienden a tener más ausencias. Los que están insatisfechos son los que tienden a faltar más al trabajo.

- Gráfica 3: En cualquiera de los departamentos, las mujeres tienden a tener mayor engagement que los hombres. En el departamento de software engineering, la diferencia es más notoria.

- Gráfica 4: El countplot indica que el personal que más se contrata son mujeres. Y vienen en su mayoría de Linkedin e Indeed.

## 5. (10 pts) Realiza una comparativa de salarios a través de diferentes grupos mediante un Faceted histogram. Responde cuál es el género y estado civil con mayor prevalencia de remuneración baja.

```python
fh = sns.FacetGrid(df, row="Sex", col="MaritalDesc", margin_titles=True, height=3, aspect=1.5)
fh.map(sns.histplot, "Salary", bins=10, color="skyblue")
fh.set_axis_labels("Salario", "Cantidad")
fh.fig.subplots_adjust(top=0.9)
fh.fig.suptitle("Comparativa de salarios por Género y Estado Civil")
plt.show()
```

El grupo con la remuneración más baja es el de hombres solteros. Hay muy pocos que están en el rando de 50k a 60k.
