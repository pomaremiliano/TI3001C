```python
import matplotlib.pyplot as plt

# Lista de colores
triada = ["#242c43", "#43242c", "#2c4324"]
analogo = ["#242c43", "#2b2443", "#243c43"]
monocromatico = [
    "#000000",
    "#101118",
    "#181c28",
    "#20273a",
    "#313951",
    "#4d546e",
    "#6a718c",
    "#888fab",
]
complementarios = ["#242c43", "#433b24"]
comp_divididos = ["#242c43", "#412443", "#384324"]
tetradico = ["#242c43", "#43243b", "#433a24", "#24432b"]

```

## Triada

```python
# Crear una figura
plt.figure(figsize=(8, 6))

# Crear un gr√°fico de barras horizontales con los colores
bar_width = 0.4  # Grosor de las barras
plt.barh(
    range(len(triada)),
    [1] * len(triada),
    color=triada,
    edgecolor="black",
    height=bar_width,
)

# Configurar los ejes
plt.yticks(range(len(triada)), triada)
plt.xticks([])
plt.title("Triada de Colores")

# A√±adir los c√≥digos de color en el centro de cada barra
for index, value in enumerate(triada):
    plt.text(0.5, index, value, ha="center", va="center", color="black", fontsize=10)

# Mostrar la gr√°fica
plt.show()
```
