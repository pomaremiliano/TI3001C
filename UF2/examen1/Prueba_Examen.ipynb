## 10ma pregunta
Se necesitan obtener los siguientes datos del DataFrame (usando 2 decimas en caso de no ser números enteros). Obtén lo solicitado, cuidando que se visualice el resultado obtenido.

* El total de existencias de productos dados de alta a partir del 2003:
* Promedio de ganancia por unidad de aquellos productos dados de alta en el 2002

* Promedio de precio de venta:
* Total de ingresos en caso de que se venda todo el inventario:

```python
#El total de existencias de productos dados de alta a partir del 2003
mayor2003 = df['fecha_alta'] >= '2003-01-01'

df.loc[mayor2003].existencia.sum()
```

```python
# Promedio de ganancia por unidad de aquellos productos dados de alta en el 2002
mayor2002 = df['fecha_alta'] >= '2002-01-01'
menor2003 = df['fecha_alta'] < '2003-01-01'
df.loc[mayor2002 & menor2003].GananciaUnidad.mean().round(2)
```

```python
# Promedio de precio de venta
df.precioVenta.mean().round(2)
```

```python
# Total de ingresos en caso de que se venda todo el inventario
(df.existencia * df.precioVenta).sum().round(2)
```
