## <a id='toc1_1_'></a>[ ¿Qué es BigData?](#toc0_)


Es una forma de transformar, recopilar, analizar y almacenar grandes cantidades de datos, centrándose principalmente en el volumen y la velocidad. Sin embargo, el Big Data también considera la variedad, variabilidad, veracidad, visualización y valor de los datos. Se utiliza principalmente en grandes empresas con volúmenes masivos de información, lo que se traduce en operaciones más eficientes, mayores ganancias y la capacidad de identificar patrones y tendencias para tomar decisiones informadas. En pocas palabras, se dice Big Data cuando la cantidad de datos ya no se pueden manejar de manera sencilla con consultas normales y archivos porque se vuelve lento e ineficiente.


## <a id='toc1_2_'></a>[¿Qué son las Bases de Datos No Relacionales / NoSQL?](#toc0_)

Son aquellas que no usan un esquema tabular de filas y columnas el cual se utiliza principlamente para consultar y almacenar datos de manera flexible.

## <a id='toc1_3_'></a>[Ventajas de trabajar con base de datos NoSQL orientadas a documentos](#toc0_)

- Escalabilidad
- Simplicidad
- Permite el intercambio de datos
- Menor costo en administración.
- Ideal para trabajar con cantidades masivas de datos.

## <a id='toc1_5_'></a>[Ejercicio 2](#toc0_)
- Investiga en la documentación el operador regex Links to an external site. y carga a un dataframe de pandas la información de películas que en su sinopsis (plot) contenga la palabra "dragon". Es decir, el equivalente a: `SELECT * FROM movies WHERE plot LIKE '%dragon%'`

```python
movies = samplemflix.movies
projection = {"title": 1, "plot" : 1}

df = pd.DataFrame(
    list(movies.find({"plot": {"$regex": " dragon ", "$options" : 'im' }}, projection))
)
df
```

## <a id='toc1_6_'></a>[Ejercicio 3](#toc0_)
- Usando la Base de Datos  sample_airbnb en MongoDB carga a un dataframe de pandas la información de propiedades que tengan al menos 30 opiniones (reviews) y sean apartamentos.

```python
sampleairbnb = client.sample_airbnb
propiedades = sampleairbnb.listingsAndReviews
projection = {"_id": 0, "name": 1, "property_type": 1, "number_of_reviews": 1}

df = pd.DataFrame(
    list(propiedades.find({"property_type": "Apartment", "number_of_reviews": {"$gte": 30}}, projection))
)
df
```

## <a id='toc1_7_'></a>[Ejercicio 4](#toc0_)
- Carga a un dataframe propiedades que tienen al menos 5 habitaciones y están en "Portugal".

```python
sampleairbnb = client.sample_airbnb
propiedades = sampleairbnb.listingsAndReviews
projection = {"_id": 0, "name": 1, "bedrooms" : 1, "address.country": 1}

df = pd.DataFrame(
    list(propiedades.find({"bedrooms" : {"$gte": 5}, "address.country" : "Portugal"}, projection))
)
df
```

## <a id='toc1_8_'></a>[Ejercicio 5](#toc0_)
- Carga a un dataframe de pandas la información de propiedades que en su resumen (summary) contenga la palabra "historic". Es decir, el equivalente a: `SELECT * FROM listingsAndReviews WHERE summary LIKE '%historic%'`

```python
sampleairbnb = client.sample_airbnb
propiedades = sampleairbnb.listingsAndReviews
projection = {"_id": 0, "name": 1, "summary" : 1}

df = pd.DataFrame(
    list(propiedades.find({"summary" : {"$regex" : " historic ", "$options": "si"}}, projection))
)
df
```
