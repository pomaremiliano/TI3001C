(20 pts) Crea una columna llamada days_till_incident donde se muestren los días que tomó desde el registro de fecha de vinculación de poliza (bind date) hasta el incidente. Reporta el promedio de estos días e indica y justifica el tipo de dato de tiempo apropiado.


**Justificación**: Salió timedelta porque es un tipo de dato que impone python cuando hay una operación de diferencia de fechas. Tal cual calcula la duración. En este caso sale 03:21:36 porque saqué el promedio. Podría ser confudo porque fue la diferencia entre 2 fechas sin tiempo.
