# Ejercicio 8
Muestra promedio, mínimo y máximo de precios de categorías de productos acompañados de nombre y descripción de categoría.

```python
conn = sqlite3.connect("shop.sqlite")
cursor = conn.cursor()

qry11 = '''
SELECT 
    c.CategoryName, 
    c.Description, 
    AVG(p.Price) AS Promedio, 
    MAX(p.Price) AS Maximo,
    MIN(p.Price) AS Minimo
FROM Products p
JOIN Categories c ON p.CategoryID = c.CategoryID
GROUP BY c.CategoryID;

'''
df_promedio = pd.read_sql_query(qry11, conn)
conn.close()
```

```python
df_promedio
```

## Ejercicio 9
Muestra fecha, cantidad, nombres de cliente y empleado de las órdenes de compra registradas por empleados desde 1997.

```python
conn = sqlite3.connect("shop.sqlite")
cursor = conn.cursor()

qry12 = '''
SELECT 
    o.OrderDate, 
    od.Quantity, 
    c.CustomerName, 
    e.FirstName
FROM Orders o, OrderDetails od, Customers c, Employees e
WHERE o.OrderID = od.OrderID
AND o.CustomerID = c.CustomerID
AND o.EmployeeID = e.EmployeeID
AND o.OrderDate >= '1997-01-01'
ORDER BY o.OrderDate ASC
'''
df_registros_empleados = pd.read_sql_query(qry12, conn)
conn.close()
```

```python
df_registros_empleados
```

## Ejercicio 10
Lista si los empleados “Davolio” o “King” han registrado más de 15 órdenes.

```python
conn = sqlite3.connect("shop.sqlite")
cursor = conn.cursor()

qry13 = '''
SELECT 
    e.FirstName,
    e.LastName,
    COUNT(o.OrderID) AS NumerodeOrdenes
FROM Employees AS e
JOIN Orders AS o 
ON e.EmployeeID = o.EmployeeID
GROUP BY e.EmployeeID
HAVING COUNT(o.OrderID) > 15 AND e.LastName IN ('Davolio', 'King')
'''
df_registros_empleados = pd.read_sql_query(qry13, conn)
conn.close()
```

```python
df_registros_empleados
```
