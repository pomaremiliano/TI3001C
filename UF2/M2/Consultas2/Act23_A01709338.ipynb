```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("shop.sqlite")
conn
```

## Ejercicio 1

(5 pts) Get customer name, order date, and shipper name for all orders. Order results using order date.


```python
df_consulta1 = pd.read_sql_query(
    """
    SELECT c.CustomerName, o.OrderDate, s.ShipperName 
    FROM Customers AS c 
    JOIN Orders AS o ON c.CustomerID = o.CustomerID 
    JOIN Shippers s ON o.ShipperID = s.ShipperID 
    ORDER BY o.OrderDate DESC
    """,
    conn,
)
df_consulta1
```

## Ejercicio 10

(10 pts) Get the top 3 best-selling products by total quantity ordered, along with their category and supplier information


```python
df_consulta10 = pd.read_sql_query(
    """
    SELECT p.ProductName, c.CategoryName, s.SupplierName, SUM(od.Quantity) as TotalQuantity
    FROM OrderDetails od
    JOIN Products p ON od.ProductID = p.ProductID
    JOIN Categories c ON p.CategoryID = c.CategoryID
    JOIN Suppliers s ON p.SupplierID = s.SupplierID
    GROUP BY p.ProductID
    ORDER BY TotalQuantity DESC
    LIMIT 3
    """,
    conn,
)
df_consulta10
```

## Ejercicio 11

(10 pts) Get all orders placed by customers in the 'USA' or 'Canada' between '1996-07-01' and '1996-07-31', along with the customer name, order date, and shipper name

```python
df_consulta11 = pd.read_sql_query(
    """
    SELECT c.CustomerName, o.OrderDate, s.ShipperName
    FROM Customers AS c
    JOIN Orders AS o ON c.CustomerID = o.CustomerID
    JOIN Shippers AS s ON o.ShipperID = s.ShipperID
    WHERE c.Country IN ('USA', 'Canada') AND o.OrderDate BETWEEN '1996-07-01' AND '1996-07-31'
    ORDER BY o.OrderDate
    """,
    conn,
)
df_consulta11
```

## Ejercicio 12
(15 pts) Get the total number of orders and total quantity ordered for each employee, but only for employees who have placed at least 10 orders and have a total quantity ordered of at least 100.

```python
df_consulta12 = pd.read_sql_query(
    """
    SELECT e.EmployeeID, e.FirstName, e.LastName, COUNT(o.OrderID) as TotalOrders, SUM(od.Quantity) as TotalQuantity
    FROM Employees AS e
    JOIN Orders AS o ON e.EmployeeID = o.EmployeeID
    JOIN OrderDetails AS od ON o.OrderID = od.OrderID
    GROUP BY e.EmployeeID
    HAVING TotalOrders >= 10 AND TotalQuantity >= 100
    """,
    conn,
)
df_consulta12
```

## Ejercicio 13
(15 pts) Get the product name, supplier name, and total quantity ordered for products with a name starting with 'C' and a price between 10 and 20, ordered by the total quantity in descending order.

```python
df_consulta13 = pd.read_sql_query(
    """
    SELECT p.ProductName, s.SupplierName, p.Price, SUM(od.Quantity) as TotalQuantity
    FROM Products AS p
    JOIN Suppliers AS s ON p.SupplierID = s.SupplierID
    JOIN OrderDetails AS od ON p.ProductID = od.ProductID
    WHERE p.ProductName LIKE 'C%' AND p.Price BETWEEN 10 AND 20
    GROUP BY p.ProductID
    ORDER BY TotalQuantity DESC
    """,
    conn,
)
df_consulta13
```
