```python
import pandas as pd
import numpy as np

class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args

    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)

    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

## Ejercicio 2 
- Prueba ahora con inner, outer, left y right joins. ¿Qué observas?


### Inner Join

```python
display('df', 'df2', "pd.merge(df, df2, how='inner')")
```

### Outer Join

```python
display('df', 'df2', "pd.merge(df, df2, how='outer')")
```

### Right Join

```python
display('df', 'df2', "pd.merge(df, df2, how='right')")
```

### Left Join

```python
display('df', 'df2', "pd.merge(df, df2, how='left')")
```

### ***Respuesta***
- El ``inner join`` me regresa el dataframe con los nombres de las columnas de ambas dataframes. No hay pérdida de datos en este caso.
- El ``outer join``  Combinan y unen las llaves de los dos dataframes. Salen muchos datos NA. Esto seguro es porque no hay coincidencias en los datos aunque tengan las llaves en común.
- El ``right join`` me regresa el dataframe con la llave de la derecha. Igual salen datos en NA. 
- El ``left join`` me regresa el dataframe con la llave de la izquierda. Igual salen datos en NAN en la columna de la derecha.
