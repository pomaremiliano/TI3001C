## [Punto 7](#Punto-7)
(5 pts) Identificar si hay columnas comunes en los dataframes empleados que permitan “enlazar” los registros en las siguientes fases.

Primero, al analizar ambos dataframes, determinamos que todas las columnas que tienen son comunes, por lo que al momento de fusionarlos es mejor utilizar merge ya que de esta manera podemos conservar todas las columnas y solo agregar las filas faltantes y sin repetirlas, por lo que join outer será nuestra estrategia para enlazar ambos dataframes.

```python
#Primero impostamos la librería numpy para poder ejecurar la clase display, que es un código que
#el profesor nos enseññó en clase, con este código podemos desplegar los resultados de merge y concat de una manera más amigable
#es por eso que utilizamos este código.
import numpy as np

class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args

    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)

    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

```python
#Este es en sí el código que utilizamos para enlazar los dataframes, preferimos la opción outer
#ya que esta forma de fusión conserva todas sus columnas en común, y agrega las filas faltantes sin repetirlas,
#por lo que de esta manera, conservamos toda la información.
df1['Fecha de Transacción'] = pd.to_datetime(df1['Fecha de Transacción'])
df2['Fecha de Transacción'] = pd.to_datetime(df2['Fecha de Transacción'], format="%d/%m/%Y")


display('df1', 'df2', "pd.merge(df1, df2, how='outer')")
```

El resultado de este punto final es nuestro dataframe completamente enlazado, así a futuro podremos analizar toda la información que nos proporcionó el socio formador de manera conjunta, y no por partes.

Finalmente es importante resaltar que esta no es la información ni el dataframe finales que estaremos utilizando, esto solo es la parte inicial, ya que todavia nos haría falta información tal como:

## [Punto 5](#Punto-5)

Detectar valores atípicos que podrían afectar el análisis para su posterior transformación (no en este avance). Ejemplo: Utilizar el método del rango intercuartílico para detectar valores que se encuentran por encima de 1.5 veces el rango intercuartílico.

```python
# Calculo de IQR
q1 = df1['Litros'].quantile(.25)
q3 = df1['Litros'].quantile(.75)
iqr = q3 - q1
print(iqr)
```

```python
# Calculo de whiskers
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
print(lower,upper)
```

```python
es_outlier_low = df2['Litros'] < lower
es_outlier_up = df2['Litros'] > upper
df2['Litros'].loc[(es_outlier_low)] # cuales son extremos (low)
```

```python
df2['Litros'].loc[(es_outlier_up)] # cuales son extremos (up)
```

```python
# Cuantos atipicos segun SDG?
valores_atipicos = es_outlier_up | es_outlier_low
valores_atipicos.sum() # cuento
```

```python
# devuelve valores tipicos segun SDG
df2[~valores_atipicos]
```

```python
df2_clean = df2[~valores_atipicos] # Creo DF con valores "limpios" para SDG
df2_clean.boxplot(['Litros'], vert=False)
```
