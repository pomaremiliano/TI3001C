```python
import pandas as pd

df = pd.read_csv('iris.csv')
```

```python
train_data = df.groupby('class').sample(n=45)
train_data
```

```python
df_prueba = df.drop(train_data.index)
```

```python
df_prueba
```

Vector de promedio por cada clase.

```python
mean_vectors = train_data.groupby('class').mean()

```

Con esta funciÃ³n clasificamos los datos de entrenamiento. La clasificacion se hace despues de calcular la minima distancia euclidiana. DespuÃ©s se aplica a cada axis del dataframe.

```python
def clasificador(row):
    distances = mean_vectors.apply(
        lambda mean: ((row[['sepal length', 'sepal width', 'petal length', 'petal width']] - mean) ** 2).sum() ** 0.5,
        axis=1
    )
    return distances.idxmin()

```

Genera una columna nueva con las clases clasificadas por la distancia mÃ¡s corta. Con apply, se implementa la funciÃ³n de clasificador sobre cada axis del df de los datos de prueba (15 restantes). Se despliegan con la clase original y la predicciÃ³n. 

```python
df_prueba['predicciÃ³n'] = df_prueba.apply(clasificador, axis=1)

df_prueba[['class', 'predicciÃ³n']]
```
