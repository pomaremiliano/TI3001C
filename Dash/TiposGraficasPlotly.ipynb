```python
#1.Gráfico Dispersión
#-------------------------------------------------
import plotly.express as px
import pandas as pd

educacion = pd.read_csv("Data/states_all.csv", usecols=["STATE", "LOCAL_REVENUE", "STATE_REVENUE"])

px.scatter(
    (educacion
    .query("STATE in ['CALIFORNIA', 'FLORIDA', 'TEXAS', 'ALASKA', 'WISCONSIN', 'WASHINGTON']") 
    ),
    x="LOCAL_REVENUE", 
    y="STATE_REVENUE",
    text="STATE",
    hover_name="STATE",    
    color="STATE",
    trendline="ols",
    trendline_scope="overall",    
    marginal_x="box",
    marginal_y="box",
    animation_frame="STATE"    
)
```

```python
#2.Gráfico Línea
#-------------------------------------------------
import plotly.express as px
import pandas as pd

educacion = pd.read_csv("Data/states_all.csv", usecols=["STATE", "YEAR", "TOTAL_REVENUE"])

px.line(
    (educacion
     .query("1992 < YEAR < 2017")
     .groupby(["YEAR"], as_index=False).sum()
    ), 
    x="YEAR", 
    y="TOTAL_REVENUE",
    title="GANANCIA EN EDUCACION A TRAVES DEL TIEMPO"
)
```

```python
#3.Gráfico de Línea - Series Múltiples
#-------------------------------------------------
import plotly.express as px
import pandas as pd

educacion = pd.read_csv("Data/states_all.csv", usecols=["STATE", "YEAR", "TOTAL_EXPENDITURE"])

px.line(
    educacion.query("2000 < YEAR < 2017 and STATE in ['CALIFORNIA', 'MASSACHUSETTS', 'WISCONSIN', 'WASHINGTON']"),
    x="YEAR",
    y="TOTAL_EXPENDITURE",
    color="STATE",
    title="Gasto Total en Educación California vs. Massachusetts",
)
```

```python
#4.Gráfico de Línea 2 - Series Múltiples
#-------------------------------------------------
educacion = pd.read_csv("Data/states_all.csv", usecols=["STATE", "YEAR", "TOTAL_REVENUE"])

px.line(
    (educacion
     .query("2000 < YEAR < 2017 and STATE in ['CALIFORNIA', 'MASSACHUSETTS']")
    ), 
    x="YEAR", 
    y="TOTAL_REVENUE",
    color="STATE",
    title="GANANCIAS EN EDUCACION A TRAVES DEL TIEMPO"
)
```

```python
#5.Gráfico de Barra
#-------------------------------------------------

educacion = pd.read_csv("Data/states_all.csv", usecols=["STATE", "YEAR", "TOTAL_REVENUE"])

px.bar(
    (educacion.query("2013 < YEAR < 2017 and STATE in ['CALIFORNIA', 'MASSACHUSETTS', 'TEXAS', 'FLORIDA']")
     #.sort_values("TOTAL_REVENUE",ascending=False)
    ),
    x="YEAR",
    y="TOTAL_REVENUE",
    color="STATE",
    labels={'YEAR': 'Año', 'TOTAL_REVENUE':'Ganancias Totales'},
    title="GANANCIAS TOTALES EN EDUCACION CALIFORNIA Y MASSACHUSETTS",
    barmode='group'
)
```

```python
#Gráfico de Burbuja
#-------------------------------------------------
educacion = pd.read_csv("Data/states_all.csv", usecols=["STATE", "AVG_MATH_8_SCORE", "AVG_READING_8_SCORE", "TOTAL_EXPENDITURE"])

px.scatter(
    (educacion
     .query("STATE in ['CALIFORNIA', 'MASSACHUSETTS', 'TEXAS', 'FLORIDA']")
     .dropna()
    ), 
    x="AVG_MATH_8_SCORE", 
    y="AVG_READING_8_SCORE",
    size="TOTAL_EXPENDITURE",
    title="CORRELACION PROMEDIOS MATEMATICAS Y LECTURA VS GASTO EN EDUCACION",
)
```

```python
#Gráfico de Pay / Dona
#-------------------------------------------------
educacion = pd.read_csv("Data/states_all.csv", usecols=["STATE", "YEAR", "GRADES_ALL_G"])

px.pie(
    educacion
    .query("YEAR == 2017 and STATE in ['CALIFORNIA', 'OREGON', 'WASHINGTON', 'IDAHO', 'WYOMING', 'NORTH_DAKOTA']")
    , 
    values="GRADES_ALL_G",
    names="STATE",
    color_discrete_sequence=px.colors.qualitative.Alphabet,
 #   hole=.4,
    title="Graduados por Estado",
    category_orders={"STATE": ["NORTH_DAKOTA", "WASHINGTON", "OREGON", "IDAHO", "WYOMING", "CALIFORNIA"]},
)
```

```python
#8.Gráfico de Histograma
#-------------------------------------------------

resultados = pd.read_csv("Data/states_all.csv", usecols=["AVG_MATH_4_SCORE"])

px.histogram(
    resultados,
    x="AVG_MATH_4_SCORE", 
    nbins=10,
    histnorm="percent",
    text_auto=True,  
    title="Resultados Matemáticas 4to Grado",
)
```
