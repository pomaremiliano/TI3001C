## Curva Cuadr√°tica


```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

# Generar 100 datos
np.random.seed(42)
X = np.linspace(-10, 10, 100).reshape(-1, 1)
y = 2 * X**2 + 3 * X + 5 + np.random.normal(0, 10, size=X.shape)  # y = 2x¬≤ + 3x + 5 + ruido

# Transformar para regresi√≥n cuadr√°tica
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Ajustar el modelo
model = LinearRegression()
model.fit(X_poly, y)

# Predicciones
y_pred = model.predict(X_poly)

# Mostrar resultados
print("Coeficientes del modelo:", model.coef_)
print("Intercepto:", model.intercept_)

# Visualizaci√≥n
plt.scatter(X, y, color='skyblue', label='Datos con ruido')
plt.plot(X, y_pred, color='red', linewidth=2, label='Modelo cuadr√°tico')
plt.title("Regresi√≥n Cuadr√°tica con 100 puntos")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

## Regresi√≥n C√∫bica

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Generar datos
np.random.seed(42)
X = np.linspace(-10, 10, 100).reshape(-1, 1)
y = 1.2 * X**3 - 0.5 * X**2 + 2 * X + 3 + np.random.normal(0, 50, size=X.shape)

# Transformar a polinomio grado 3
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)

# Ajustar modelo
model = LinearRegression()
model.fit(X_poly, y)

# Predicciones
y_pred = model.predict(X_poly)

# Mostrar coeficientes
print("Coeficientes del modelo:", model.coef_)
print("Intercepto:", model.intercept_)

# Visualizaci√≥n
plt.scatter(X, y, color='lightblue', label='Datos con ruido')
plt.plot(X, y_pred, color='darkred', linewidth=2, label='Modelo c√∫bico')
plt.title("Regresi√≥n C√∫bica con 100 puntos")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

## Regresi√≥n Exponencial

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# 1. Generar datos
np.random. seed(0)
X = np.linspace(0, 4, 100).reshape(-1, 1)
y = 2 * np.exp(0.8 * X) + np.random.normal(0, 1.5, size=X.shape)

# 2. Asegurar que y sea positiva
y = np.clip(y, a_min=0.1, a_max=None)

# 3. Transformar y con logaritmo
y_log = np.log(y)

# 4. Ajustar regresi√≥n lineal
model = LinearRegression()
model.fit(X, y_log)

# 5. Obtener coeficientes
b = model.coef_[0] [0]
ln_a = model.intercept_[0]
a = np.exp(ln_a)

print(f"Modelo ajustado: y = {a:.2f} * exp({b:.2f} * x)")

# 6. Predicciones
y_pred = a * np.exp(b * X)

# 7. Calcular R2 y MSE
r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)

print(f"R2 (coeficiente de determinaci√≥n) : {r2 :.4f}")
print(f"Error cuadr√°tico medio (MSE) : {mse :.4f}")

# 8. Visualizaci√≥n
plt.scatter(X, y, color='lightcoral', label='Datos reales')
plt.plot(X, y_pred, color='darkred', linewidth=2, label='Modelo exponencial ajustado')
plt.title("Regresi√≥n Exponencial (100 puntos)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
