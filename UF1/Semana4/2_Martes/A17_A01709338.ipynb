```python
class Profesor:

    def __init__(self, nomina, nombre, edad):
        self.__nomina = nomina
        self.__nombre = nombre
        self.__edad = edad


    def pagoMensual():
        pass

    def getNomina(self):
        return self.__nomina

    def getNombre(self):
        return self.__nombre

    def getEdad(self):
        return self.__edad

    def setNomina(self, nomina):
        self.__nomina = nomina

    def setNombre(self, nombre):
        self.__nombre = nombre

    def setEdad(self, edad):
        self.__edad = edad

    def mostrar(self):


        return (
            "Nomina: "
            + str(self.getNomina())
            + "\n"
            + "Nombre: "
            + self.getNombre()
            + "\n"
            + "Edad: "
            + str(self.getEdad())
            + "\n"
        )
```

```python
class Planta(Profesor):
    def __init__(self, nomina, nombre, edad, salarioMensual):
        super().__init__(nomina, nombre, edad)
        self.__salarioMensual = salarioMensual

    def pagoMensual(self, beneficios):
        self.beneficios = beneficios
        return self.__salarioMensual + self.beneficios

    def getSalarioMensual(self):
        return self.__salarioMensual

    def setSalarioMensual(self, salarioMensual):
        self.__salarioMensual = salarioMensual

    def mostrar(self):
        return (
            "Nomina: "
            + self.getNomina()
            + "\n"
            + "Nombre: "
            + self.getNombre()
            + "\n"
            + "Edad: "
            + str(self.getEdad())
            + "\n"
            + "Salario Mensual: "
            + str(self.getSalarioMensual())
            + "\n"
        )
```

```python
class Catedra(Profesor):
    def __init__(self, nomina, nombre, edad, horasTrabajadas, pagoHora):
        super().__init__(nomina, nombre, edad)
        self.__horasTrabajadas = horasTrabajadas
        self.__pagoHora = pagoHora

    def pagoMensual(self, beneficios):
        self.beneficios = beneficios
        return self.__horasTrabajadas * self.__pagoHora + self.beneficios

    def getHorasTrabajadas(self):
        return self.__horasTrabajadas

    def getPagoHora(self):
        return self.__pagoHora

    def setHorasTrabajadas(self, horasTrabajadas):
        self.__horasTrabajadas = horasTrabajadas

    def setPagoHora(self, pagoHora):
        self.__pagoHora = pagoHora

    def mostrar(self):
        return (
            "Nomina: "
            + self.getNomina()
            + "\n"
            + "Nombre: "
            + self.getNombre()
            + "\n"
            + "Edad: "
            + str(self.getEdad())
            + "\n"
            + "Horas Trabajadas: "
            + str(self.getHorasTrabajadas())
            + "\n"
            + "Pago por Hora: "
            + str(self.getPagoHora())
            + "\n"
        )
```

```python
def main():
    t = Profesor("A076545678", "Juan", 30)
    print("--- PROFESOR ---\n")
    print(t.getNomina())
    print(t.getNombre())
    print(t.getEdad())
    print(t.mostrar())
    nueva_nomina = input("Ingrese nueva nomina: ")
    t.setNomina(nueva_nomina)
    nuevo_nombre = input("Ingrese nuevo nombre: ")
    t.setNombre(nuevo_nombre)
    nueva_edad = input("Ingrese nueva edad: ")
    t.setEdad(nueva_edad)
    print(t.mostrar())

    p = Planta("A76543", "Ivan", 23, 48700)
    print("--- PROFESOR DE PLANTA ---\n")
    print("Salario Mensual Profesor Planta: ", p.getSalarioMensual())
    print(p.mostrar())
    nuevo_salario_mensual = float(input("Ingrese nuevo salario mensual: "))
    p.setSalarioMensual(nuevo_salario_mensual)
    print(p.mostrar())
    pagomensualp = p.pagoMensual(600)
    print("Pago Mensual Profesor Planta (600): ", pagomensualp)

    c = Catedra("A876543", "Carlos", 71, 30, 3000)
    print("\n --- PROFESOR DE CATEDRA ---\n")
    print("Horas Trabajadas Profesor Catedra: ", c.getHorasTrabajadas())
    print("Pago por Hora Profesor Catedra: ", c.getPagoHora())

    print(c.mostrar())
    nuevas_horas_trabajadas = float(input("Ingrese las nuevas horas trabajadas: "))
    c.setHorasTrabajadas(nuevas_horas_trabajadas)
    nuevo_pagoporhora = float(input("Ingrese el nuevo pago por hora: "))
    c.setPagoHora(nuevo_pagoporhora)
    print(c.mostrar())

    pagomensualc = c.pagoMensual(900)
    print("Pago Mensual Profesor Catedra: ", pagomensualc)

    lista_profesores = [t, p, c]
    for profesor in lista_profesores:
        print("---- Lista de Profesores ----\n", profesor.mostrar())


main()
```
