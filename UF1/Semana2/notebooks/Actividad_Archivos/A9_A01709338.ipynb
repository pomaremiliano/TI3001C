## Ejercicio 1

La función escribe_en_archivo (nombre) que recibe el nombre del archivo de texto. La función deberá pedir al usuario cuatro oraciones o frases y las agregará en el archivo de texto.


```python
def escribe_en_archivo(nombre):
    file = open(nombre, "w")
    for i in range(1, 5):
        frase = input("Ingresa frase : ")
        file.write(frase + "\n")
    file.close()


def main():
    nombre = input("Ingresa nombre de archivo: ")
    escribe_en_archivo(nombre)


main()
```

## Ejercicio 2

La función cuenta_digitos (nombre) que recibe el nombre del archivo de texto y regrese el número de dígitos contenidos en un archivo de texto.

Por ejemplo: Si en el archivo de texto tiene almacenado el texto:

"Luz" 1973
_Zoila_ 2000
-Pedro- 2024


```python
def cuenta_digitos(nombre):
    digitos = "0123456789"
    file = open(nombre, "r")
    contenido = file.read(1)
    cont = 0

    while contenido != "":
        if contenido in digitos:
            cont += 1
        contenido = file.read(1)
    file.close()
    return cont


def main():
    nombre_archivo = input("Ingresa el nombre del archivo: ")
    total_digitos = cuenta_digitos(nombre_archivo)
    print("El archivo %s contiene %s digitos." % (nombre_archivo, total_digitos))


main()
```

## Ejercicio 3

La función cuenta_caracteres (nombre) que recibe el nombre del archivo de texto y regrese el número de caracteres contenidos en el archivo de texto, sin incluir espacios en blanco y saltos de línea.


```python
def cuenta_caracteres(nombre):
    file = open(nombre, "r")
    contenido = file.read(1)
    cont = 0

    while contenido != "":
        if contenido != " " and contenido != "\n":
            cont += 1
        contenido = file.read(1)
    file.close()
    return cont


def main():
    nombre_archivo = input("Ingresa el nombre del archivo: ")
    total_caracteres = cuenta_caracteres(nombre_archivo)
    print("El archivo %s contiene %s caracteres" % (nombre_archivo, total_caracteres))


main()
```
