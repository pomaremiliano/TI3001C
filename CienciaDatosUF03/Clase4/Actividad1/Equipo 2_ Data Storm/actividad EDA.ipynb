```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Highest Holywood Grossing Movies.csv")
df.info()
```

```python
df
```

```python
df['Budget (in $)'] = pd.to_numeric(df['Budget (in $)'].str.replace('[\$,]', '', regex=True), errors='coerce')
df['Domestic Opening (in $)'] = pd.to_numeric(df['Domestic Opening (in $)'].str.replace('[\$,]', '', regex=True), errors='coerce')
```

```python
mean_values = df.mean(numeric_only=True)
print(mean_values)
```

```python
median_values = df.median(numeric_only=True)
print(median_values)
```

```python
mode_values = df.mode().iloc[0]
print(mode_values)
```

```python
std_values = df.std(numeric_only=True)
print(std_values)
```

```python
cv_values = (std_values / mean_values) * 100
print(cv_values)
```

```python
quantiles = df.quantile([0.25, 0.5, 0.75], numeric_only=True)
quantiles
```

## Para este EDA, nos concentraremos principalmente en la distribución de películas más taquilleras a través de los años. Para esto, primero lo acomodaremos en décadas usando bins.

```python
bins = list(range(df['Year'].min(), df['Year'].max() + 1, 10))
out = pd.cut(df['Year'], bins=bins)
out.value_counts()
```

- Aquí podemos observar que el rango de años en los que más han habido películas taquilleras es entre 2007 y 2017.
