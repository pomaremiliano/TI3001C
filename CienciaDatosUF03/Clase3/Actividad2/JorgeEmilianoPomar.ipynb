```python
import pandas as pd 

df = pd.read_csv("sales_data_sample-2.csv")
```

## Gráfica comparativa por cuarto del año

```python
import seaborn as sns

print(df.columns)

df.value_counts("QTR_ID")
df.value_counts("MONTH_ID")

df.value_counts("QTR_ID").sort_values(ascending=False).plot(kind="bar", figsize=(10, 5), title="QTR_ID")
sns.countplot(x=df["QTR_ID"])
```

## Gráfica comparativa por mes

```python
df.value_counts("MONTH_ID").sort_values(ascending=False).plot(kind="bar", figsize=(10, 5), title="MONTH_ID")
sns.countplot(x=df["MONTH_ID"])
```

Considero que la gráfica ordenada por meses puede ser útil para un análisis que se enfoque en solamente en un mes en particular. Pero creo que es más útil la gráfica que ordena por cuartos, ya que es más fácil de comparar e interpretar con intervalos o categorías más grandes. Probablemente, lo ideal es primero observar la gráfica de cuertos, en la cuál te dirá probablemente el estado de las ventas. Y la gráfica de meses te dirá un detalle más específico de un mes en particular.
